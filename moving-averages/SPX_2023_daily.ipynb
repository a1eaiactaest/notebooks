```python
import mplfinance as mpf
import pandas as pd
```

```python
#* Get data from Stooq 2022 daily chart
data = pd.read_csv("https://stooq.com/q/d/l/?s=^spx&d1=20230101&d2=20231231&i=d",index_col=0,parse_dates=True)
data.index.name = 'Date'
# data.shape
# data.head(3)
# data.tail()
data.head()
```

```python
_, axes = mpf.plot(
    data, type="line", style="yahoo", mav=(8, 20, 50), returnfig=True
)

axes[0].legend([None] * 5)
handles = axes[0].get_legend().legendHandles
axes[0].legend(handles=handles, labels=list(["Close Price", "MA 8", "MA 20", "MA 50"]))
```

```python
mpf.plot(data,type='candle',mav=(8,20,50),style="yahoo")
```
