```python
import pandas as pd
import mplfinance as mpf
```

```python
#* Get data from Stooq
daily = pd.read_csv("https://stooq.com/q/d/l/?s=btc.v&d1=20230101&d2=20231231&i=d",index_col=0,parse_dates=True)
```

```python
daily.index.name = 'Date'
daily.shape
daily.head(3)
daily.tail()

mpf.plot(daily,type='ohlc',mav=(8,20,50),volume=True,style="yahoo")
```
