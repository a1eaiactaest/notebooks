```python
import numpy as np
import pandas as pd
from fracdiff import fdiff
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('src.csv')
```

```python
date = [x for x in range(len(df))][10:]
data = np.array(df.pop("Close"))
```

```python
fig, ax = plt.subplots()
ax.plot(date, data[10:], label="Close", color="lightblue")
ax.plot(date, fdiff(data, 0.2)[10:], label="FDiff 0.2", color="lightgreen")
ax.plot(date, fdiff(data, 0.5)[10:], label="FDiff 0.5", color="yellow")
ax.plot(date, fdiff(data, 0.8)[10:], label="FDiff 0.8", color="orange")
ax.legend()
ax.set_title("FracDiff effect on Close")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(date, fdiff(data, 1.2)[10:], label="FDiff 1.2", color="lightgreen")
ax.plot(date, fdiff(data, 1.5)[10:], label="FDiff 1.5", color="yellow")
ax.plot(date, fdiff(data, 1.8)[10:], label="FDiff 1.8", color="orange")
ax.legend()
ax.set_title("FracDiff effect on Close")
plt.show()
```
